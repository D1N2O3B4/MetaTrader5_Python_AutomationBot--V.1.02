```python
import MetaTrader5 as mt5
from datetime import datetime
import pandas as pd
```

```python
mt5.initialize()
```

```python
from_date = datetime(2022,1,1)
to_date = datetime(2022,2,2)

tick_range = mt5.copy_ticks_range("EURUSD",from_date,to_date,mt5.COPY_TICKS_ALL)

df_tick_range = pd.DataFrame(tick_range)

df_tick_range["time"] = pd.to_datetime(df_tick_range["time"],unit="s")

df_tick_range
```
